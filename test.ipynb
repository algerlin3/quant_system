```python
import yfinance as yf
```

```python
symbol="AAPL"
start="2020-01-01"
end="2025-12-01"
interval = "1d"
df = yf.download(
        symbol,
        start=start,
        end=end,
        interval=interval,
        auto_adjust=False,
        progress=False,
        multi_level_index=False
    )
print(df.head())
```
